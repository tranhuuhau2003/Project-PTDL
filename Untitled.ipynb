```python
import findspark
findspark.init()
import pyspark
from pyspark.sql import SparkSession
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px

#Use to hide unimportant or unnecessary warnings
import warnings
warnings.filterwarnings("ignore")

spark = SparkSession.builder.appName("Project").getOrCreate()

path = "C:\\Users\\tranh\\PTDL_disk(C)\\project\\"

df = pd.read_csv(path + "netflix_titles.csv")
df = df.dropna()
df.head()
```

```python
# Assuming df is your DataFrame and 'date_added' is the column containing the dates in string format
#remove leading and trailing whitespace
df['date_added'] = df['date_added'].str.strip()  

#convert values in column 'date_added'
df['date_added'] = pd.to_datetime(df['date_added'], format='%B %d, %Y')

#changing the month name to month number (1 to 12)
df['month'] = df['date_added'].dt.month
```

****LINE PLOT****

***Number of Titles Over the Years***

```python
#extract year from 'date_added'
df['year_added'] = df['date_added'].dt.year

#count the number of titles added each year
title_counts = df.groupby('year_added')['title'].count()

#plot the trend
plt.figure(figsize=(15, 6))
title_counts.plot(kind='line')
plt.title('Number of Titles Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()
```

****Seasonality of Movie and TV Shows Releases on Netflix****

```python
#extract month and year from the 'date_added'
df['month_added'] = df['date_added'].dt.month
df['year_added'] = df['date_added'].dt.year

#filter dataframe
tv_shows_df = df[df['type'] == 'TV Show']
movies_df = df[df['type'] == 'Movie']

#group by month and year(groupby()); count the number of releases(size()), and calculate the total for each genre
tv_shows_monthly_counts = tv_shows_df.groupby(['month_added', 'year_added']).size().unstack(fill_value=0)
movies_monthly_counts = movies_df.groupby(['month_added', 'year_added']).size().unstack(fill_value=0)

#seasonality for TV shows
plt.figure(figsize=(14, 6))
tv_shows_monthly_counts.mean(axis=1).plot(label='TV Shows', marker='o')
plt.title('Seasonality of TV Show')
plt.xlabel('Month')
plt.ylabel('Average Number of Releases')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.show()

#seasonality for movies
plt.figure(figsize=(14, 6))
movies_monthly_counts.mean(axis=1).plot(label='Movies', marker='o')
plt.title('Seasonality of Movie')
plt.xlabel('Month')
plt.ylabel('Average Number of Releases')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.show()
```

****Histogram Plot****

****Distribution of Movie Durations on Netflix****

```python
#filter dataframe 
movies_df = df[df['type'] == 'Movie']

#extract the duration values and convert them to integers
movie_durations = movies_df['duration'].str.replace(' min', '')
    
#replace the string '1 Season' with the NaN value
movie_durations = movie_durations.replace('1 Season', np.nan).dropna().astype(int)

plt.figure(figsize=(15, 6))
plt.hist(movie_durations, bins=20, edgecolor='black', color='green')
plt.title('Distribution of Movie Durations on Netflix')
plt.xlabel('Duration')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)  
plt.show()
```

```python
# Xử lý sơ bộ cột "rating" (xử lý lỗi tiềm ẩn)
try:
    df["rating_numeric"] = pd.to_numeric(df["rating"], errors='coerce')
except:
    print("Cảnh báo: Gặp các giá trị xếp hạng không phải số. Giá trị NaN sẽ được sử dụng cho chúng.")
    df["rating_numeric"] = pd.to_numeric(df["rating"], errors='coerce')

# Xóa các hàng có xếp hạng số bị thiếu (tùy chọn)
# df.dropna(subset=['rating_numeric'], inplace=True)  # Bỏ ghi chú để xóa các hàng có NaN

# Tạo biểu đồ phân tán với đường hồi quy (tất cả dữ liệu)
sns.lmplot(x="rating_numeric", y="release_year", data=df, fit_reg=True)
plt.title("Xếp hạng so với Năm phát hành (Tất cả các phim Netflix)")
plt.xlabel("Xếp hạng (Số)")
plt.ylabel("Năm phát hành")
plt.grid(True)
plt.show()

# Khám phá theo loại (Phim so với Chương trình truyền hình)
if "type" in df.columns:
    sns.lmplot(x="rating_numeric", y="release_year", data=df, hue="type", fit_reg=True)
    plt.title("Xếp hạng so với Năm phát hành theo Loại (Phim Netflix)")
    plt.xlabel("Xếp hạng (Số)")
    plt.ylabel("Năm phát hành")
    plt.grid(True)
    plt.show()
```
